```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline

import scipy.fftpack as fft
```

```python
pic = plt.imread('cnoisy.jpg')
```

```python
#plt.figure(figsize=(12,9))
plt.imshow(pic,cmap='gray')
```

```python
pic_data = fft.fft2(pic)
```

```python
pic_data.min()
```

```python
pic_fft_2 = np.where(np.abs(pic_data)>5e3,0,pic_data)
```

```python
pic_ifft = fft.ifft2(pic_fft_2)
```

```python
pic_ifft_real = np.real(pic_ifft)
```

```python
pic_ifft_real
```

```python
plt.show(pic_ifft_real)
```
