```python
import numpy as np
import pandas as pd
import sympy as sym
import matplotlib
import matplotlib.pyplot as plt
#from matplotlib import rcParams

%matplotlib inline
```

```python
x = np.linspace(-np.pi,np.pi,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.title('测试中文 $f(x) = sin(x) $',fontsize=20,loc = 'center')
plt.xticks([-np.pi,0,np.pi],['$-\pi$',0,'$\pi$'])
plt.ylabel('$f(x) = sin(x) $',size=20,rotation=90)
plt.legend(['sin(x)','cos(x)'])
plt.annotate(s = 'Origin',xy=(0,0),xytext=(1,-1),arrowprops=
             {'width':2,'headwidth':12,'headlength':24,'shrink':0.01})
```
