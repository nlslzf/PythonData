### 保存图片

## 二、设置plot的风格和样式
plot语句中支持除X，Y以外的参数，以字符串形式存在，来控制颜色、线型、点型等要素，语法形式为：
plt.plot(X,Y,'format',...)
### 点和线的样式
1. 颜色  color    
2. 透明度 alpha    
3. 线型   
'-'实线 '--'破折线 '-.'点划线 ':'虚线 'steps'阶梯线  'None'/','什么都不画     
4. 线宽 linewidth或lw参数    
5. dashes 线的长度和间隔    
6. marker  
'1'一角朝下的三脚架  '2'一角朝上的三角架  '3'一角朝左的三角架  '4'一角朝右的三角架  's'正方形  'h'六边形1  '8'八边形  'p'五边形 'H'六边形2  '.'点 '\*'星号 ','像素 'x'X  '+'加号 'v'一角朝下的三角形 '<'一角朝左的三角形 '>'一角朝右的三角形

```python
x = np.arange(0,10,0.1)
plt.plot(x,np.sin(x),'r',alpha=0.5,marker='*',markersize=10)
```

## 四、图形内的文字、注释、箭头
控制文字属性的方法：

|Pyplot函数|API方法|描述|
|:----:|:----:|:----:|
|text()|mpl.axes.Ases.test()|在Axes对象的任意位置添加文字|
|xlabel()|mpl.axes.Ases.set_xlabel()|为x轴添加标签|
|ylabel()|mpl.axes.Ases.set_ylabel()|为y轴添加标签|
|title()|mpl.axes.Ases.set_title()|为Axes对象添加标题|
|legend()|mpl.axes.Ases.legend()|为Axes对象添加图例|
|figtext()|mpl.figure.Figure.legend()|在Figure对象的任意位置添加文字|
|suptitle()|mpl.figure.Figure.suptitle()|为Figure对象添加中心化的标题|
|annotate()|mpl.axes.Axes.annotate()|为Axes对象添加注释（箭头可选）|

```python
x = np.arange(0,2*np.pi,np.pi/360)
plt.plot(x,np.sin(x))
plt.text(0,0,'sin(0)=0')
plt.figtext(0.5,0.5,'sin($\pi$)=0')
```

```python
x =np.random.randint(0,20,10)
x[5] = 30
plt.figure(figsize=(8,6))
plt.plot(x)
plt.ylim([-2,35])
plt.annotate(s = 'this point is important',xy=(5,30),xytext=(7,31),arrowprops=
             {'width':2,'headwidth':12,'headlength':24,'shrink':0.05})
```

```python
plt.figure(figsize=(10,8))
plt.axis([0,10,0,20])
arrstyles=['-','->','-[','<-','<->','fancy','simple','wedge']
for i,style in enumerate(arrstyles):
    plt.annotate(style,xytext=(1,2+2*i),xy=(4,1+2*i),arrowprops=dict(arrowstyle=style))

connstyles=['arc','arc,angleA=10,armA=30,rad=30','arc3,rad=.2','arc3,rad=-.2','angle','angle3']
for i,style in enumerate(connstyles):
    plt.annotate(style,xytext=(6,2+2*i),xy=(8,1+2*i),arrowprops=dict(arrowstyle='->',connectionstyle=style))
#plt.show()
```

```python
x = np.linspace(0,8,101)
y = np.linspace(0,8,101)

# plt.xlim([0,8])
# plt.ylim([0,8])
X,Y = np.meshgrid(x,y)

def mk_z(X,Y):
    return 3+0.5*np.sin(X)*np.cos(Y)

Z = mk_z(X,Y)

```

```python
plt.figure(figsize=(10,16))
axes = plt.subplot(211,projection='3d')
axes.plot_surface(X,Y,Z)

axes2 = plt.subplot(212,projection='3d')
s = axes2.plot_surface(X,Y,Z,cmap='rainbow')
plt.colorbar(s,shrink=0.5)
```
