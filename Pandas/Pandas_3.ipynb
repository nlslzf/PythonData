```python
df = pd.read_csv('test.csv')
df2 = pd.read_csv('test2.csv')
df2
```

```python
pop = pd.merge(df,df2,left_on='姓名',right_on='name',how='outer')
display(pop)
pop.shape
```

```python
pop2 = pop.drop('name',axis=1)
pop2
```
